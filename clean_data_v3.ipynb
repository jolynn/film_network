```python
#helper functions
def is_tv(title):
    """
    Given a title, return whether the title is a TV show.
    
    Two ways to determine whether the title is TV:
    (1) The word '(TV)' in the title
    (2) The title includes season and espisode number (#n.n)
    (3) The title contains (yyyy-mm-dd)
    (4) The title contians {episode name}
    """
    if '(TV)' in title:
        return True
    if '{' in title and '}' in title and re.search(r'#[0-9][0-9]*.[0-9][0-9]*', title):
        return True
    if re.search(r'\{\([0-9]{4}-[0-9]{2}-[0-9]{2}\)\}', title):
        return True
    if re.search(r'\([0-9]{4}\)\s\{(.+)\}', title):
        return True
    return False
     
def get_year(title):
    """
    Given a title, return the year of production (int). 
    If year of production is missing, return 0.
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return int(re.search(r'\d{4}', year.group(0)).group(0))
    if re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title):
        return 0
    raise Exception, 'unknown year format: '+title
    
def get_year_id(title):
    """
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return re.search(r'\d{4}[/]*[A-Z]*', year.group(0)).group(0)
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return re.search(r'[\?]{4}[/]*[A-Z]*', missing.group(0)).group(0)
    raise Exception, 'unknown year format: '+title
         
def get_name(title):
    """
    Given a title, return name of movie/TV.
    """
    
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return title.split(year.group(0))[0]
    
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return title.split(missing.group(0))[0]
    
    raise Exception, 'unknown name format: '+title    
```

```python
def read_complete(f):
    """
    Return a dataframe of works that have complete crew records.
    """
    
    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    title_ls = []
    
    title = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        assert len(segs) >= 2, 'wrong format: '+line
        title = segs[0]
        title_ls.append(title)

                
    #save data to a dataframe
    df = pd.DataFrame({'title':title_ls})
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['year_id'] = df.title.map(get_year_id)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df  
```

```python
def read_country(f):
    """
    """
    
    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    title_ls = []
    country_ls = []
    
    title = None
    country = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        assert len(segs) >= 2, 'wrong format: '+line
        title = segs[0]
        country = segs[-1]
        title_ls.append(title)
        country_ls.append(country)
                
    #save data to a dataframe
    df = pd.DataFrame({'title':title_ls, 'country':country_ls}, columns=['title', 'country'])
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['year_id'] = df.title.map(get_year_id)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df
```

```python
def read_data(f, role):
    """
    Given a file path and crew role name, 
    return a dataframe that reads the data set in the file path. 
    The dataframe contains the following fields:
        person: crew member's name
        title: messy title from the dataset
        role: crew member's role 
        name: cleaned title of movie/tv
        year: year of production
        tv: is TV or not
        vm: is video movie or not
        vg: is video game or not
          
    """

    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    person_ls = []
    title_ls = []
    
    person = None
    title = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        if len(segs) >= 2:
            person = segs[0]
            title = segs[-1]
            person_ls.append(person)
            title_ls.append(title)

        elif len(segs) == 1:
            if len(segs[0]) != 0:
                title = segs[0]
                person_ls.append(person)
                title_ls.append(title)
            
        else:
            print segs  
    
    #save data to a dataframe
    df = pd.DataFrame({'person':person_ls, 'title':title_ls}, columns=['person', 'title'])
    df['role'] = role
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['year_id'] = df.title.map(get_year_id)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df
```

```python
def add(movie_df, crew_df, net, role):
    """
    Add nodes and edges to the bipartite network 
    Return network.
    """
    
    movie_df['new_name'] = movie_df.apply(lambda row: row['name'].strip('"').replace(' ', '_')+'_(%s)'%row['year_id'], axis=1)
    merged = movie_df.merge(crew_df[['name', 'year_id', 'person', 'role']], how='left', on=['name', 'year_id'])
    valid_merged = merged[~merged.person.isnull()]

    #add crew members 
    crews = list(set(valid_merged.person.tolist()))
    net.add_nodes_from(crews, bipartite=0, role=role)
    
    #add movies
    unique_movies_df = valid_merged.drop_duplicates(subset=['name', 'year_id'])
    net.add_nodes_from(unique_movies_df.new_name.tolist(), bipartite=1, year=unique_movies_df.year.tolist())
    
    #add edges
    edges = [(row['new_name'], row['person']) for i, row in valid_merged.iterrows()]
    net.add_edges_from(edges)
    #valid_merged.apply(lambda row: net.add_edge((row['new_name'], row['person'])), axis=1)

    return net
```

##Coreness

Source: https://networkx.lanl.gov/trac/ticket/419

```python
from scipy import optimize
from scipy.stats.stats import pearsonr
import numpy
import networkx as nx

def core_correlation(A,C):
	"""
	returns the pearson correlation between A and
	the ideal coreness matrix created from C
	
	A: ajacency matrix (valued or unvalued)
	C: 1D matrix representing the coreness of each node
	"""
	cMat = numpy.matrix(C)
	Cij = numpy.multiply(cMat,cMat.transpose())
	return pearsonr(A.flat,Cij.flat)
	
def _core_fitness(C,*args):
	"""
	converts coreCorrelation(A,C) to something useable
	with scipy.optimize (which aims to MINIMIZE a function)
	Need to express highest positive correlation as function
	to be minimized
	"""
	return core_correlation(args[0],C)[0] * -1.0

def coreness(G,return_correlation=False):
	"""
	Calculates each node's coreness value, as described by Stephen Borgatti
	
	Coreness describes to what degree a node is a member of the graph's core

	Parameters
	----------
	G : graph
		A networkx graph
	
	return_correlation : whether to return the final correlation to the ideal core / periphery structure

	Returns
	-------
	census : dictionary
			 Dictionary with nodes as keys and coreness as values
			 *if return_correlation is set, then returns a tuple (dict with nodes as keys and coreness as values,correlation)* 

	Refrences
	---------	
	.. [1] Models of Core/Periphery Structures
	   Stephen P. Borgatti, Boston College
	   http://dx.doi.org/10.1016/S0378-8733(99)00019-2
	"""
	
	A = nx.to_numpy_matrix(G)
	
	# need a starting point for the optimizer, for now using a random starting point.
	initialC = numpy.random.rand(len(A)) # can we do better? Is it important? Maybe use constraint or centrality? 
	
	# run a bfgs optimizers that optimizes correlation between calculated coreness scores and the ideal model
	best = optimize.fmin_l_bfgs_b(_core_fitness, initialC,args=(A,None),approx_grad=True,bounds=[(0.0,1.0) for i in xrange(len(A))])	
	
	part = {}
	for node in G:
		part[node] = best[0][G.nodes().index(node)]
	
	# return correlation to ideal if return_correlation is set
	if return_correlation: 
		return part,best[1] * -1.0
	return part
```

```python
"""

Original code written by Diederik van Liere and Jasper Voskuilen 
from RSM Erasmus University, the Netherlands and contributen to Jung 1.x 
This code was ported from JUNG 2.0 (jung.sourceforge.net) and extended by Diederik van Liere from RSM
Erasmus University, the Netherlands and the Rotman School of Management, University of Toronto.

This module contains the structural hole measures as suggested by Burt, 1992:
- Effective size
- Efficiency
- Network constraint
- Hierarchy

In addition it also calculates ego-density. 

Reference: Burt, Ronald S. (1992) Structural Holes - The Social Structure of Competition, Cambridge, MA: Harvard University Press
http://books.google.ca/books?hl=en&lr=&id=E6v0cVy8hVIC&oi=fnd&pg=PR7&dq=%22Burt%22+%22Structural+Holes%22+&ots=omMOXb-aPF&sig=fZL_7Ly4N9h805E59QbAXJatkx8#PPR9,M1
"""

import math
import networkx as nx

try:
    import psycho
    psyco.full()
except:
    pass

"""
g=nx.DiGraph()
# nodes are automatically added with edges
g.add_edge("node1","node2")
g.add_edge("node3","node2")
g.add_edge("node2","node1")
g.add_edge("node1","node3")
g.add_node("node4")
"""

def ego_graph(G,n,radius=1):
    """Returns induced subgraph centered at node n with given radius"""
    # FIXME: implement arbitrary radius ego graph 
    # this works for radius 1
    nodes=[n]  # add center node
    nodes.extend(G.neighbors(n)) # extend with neighbors
    return G.subgraph(nodes)

def ego_density(G,v):
    """
    Returns the density of the ego network of v. Calculated as the sum of ties between the alters of v
    excluding the ties of v divided by the number of potential ties calculated as n*(n-1) (excluding ego). 
    Value ranges between 0 and 100.
    """ 
    if G.degree(v) <= 1:
        return "."
    result = 0.0
    egosize=ego_size(G)
    result = len(egosize.get(v))
    result = float(result)
    alters_i = egosize.get(v)
    edges =0 
    for i in alters_i:
        alters_j = egosize.get(i)
        for j in alters_j:
            if (j!=v): #and G.get_edge(i,j)
                try:
                    G.get_edge(v,j)             #make sure that the alter of the alter is connected to node V
                    edges += G.get_edge(i,j)
                except:
                    pass
    if edges == 0:
        # Technically, the result is undefined (NaN), however 0.0 eases interpretation instead of a missing value.
        return result
    else: 
        result = (edges/(result*(result-1)))*100
    return result

def ego_size(G):
    """
    Creates a dictionary with as key the node and as values a list of alters (both successors and predecessors but all values are unique)
    this can be used to retrieve the size of ego's network. 
    Basically, it makes a unique dictionary of the alters using the predecessors_iter and successors_iter methods 
    and can be seen as the neighborhood implementation for directed graphs. 
    """ 
    ego = {}
    for n in G:
        alters = []
        #if G.directed == True:
        #    for i in G.predecessors_iter(n):
        #        if (i!= n):
        #            if alters.count(i) == 0: 
        #                    alters.append(i)
        #            ego[n] = alters
        #            
        #    for i in G.successors_iter(n):
        #        if (i!= n):
        #            if alters.count(i) == 0:
        #                    alters.append(i)
        #            ego[n] = alters
        #else:
        for i in G.neighbors_iter(n):
            if (i!=n):
                alters.append(i)
            ego[n] = alters

        
    return ego
        

def effective_size_aric(G,v):
    """
    Diederik to Aric: I have not fixed your code but left it here.
    FIXME: The starting point for a nodes' effective size is not degree
    but the size of the ego network.  
    
    E=ego_graph(G,v)
    vdeg=E.degree(v)
    deg=[k-1.0 for k in E.degree(E[v])]
    return vdeg-sum(deg)/vdeg
    """

def effective_size(G,v):
    """
    Burts's effective size measure for directed and weighted graphs.Essentially, the
     * number of neighbors minus the average degree of those in v's neighbor set,
     * not counting ties to v. 
    This is my version of effective size and I have validated it against UCINET and it seems to work.
    that is: so far :), my results match the results of UCINET. 
    """
    result = 0.0
    if G.degree(v)==0: 
        return result
    egosize=ego_size(G)
    result = len(egosize.get(v))
    alters_i = egosize.get(v)
    for u in alters_i:
        alters_j = egosize.get(u)
        for w in alters_j:
            if ((w != v) and ( w!= u)):
                result -= normalized_mutual_edge_weight(G,v,w) * max_scaled_mutual_edge_weight(G,u,w)      
                    
    return result

def efficiency(G,v): 
    return effective_size(G,v)/G.degree(v) 

def borgatti_effective_size(G,v):
    """
    Aric: is this the same as effective size? 
    Diederik: Yes, but this is an 'easier' implementation as it does not consider edge weights nor the direction of the edges.
    This can be used to benchmark Burt's effective size for undirected / unweighted graphs. 
    Reference: Borgatti, Stephen P. (1997) "Unpacking Burt's Redundancy Measures", Connections 20(1):35-38
    http://www.analytictech.com/connections/v20(1)/holes.htm
    """
    borgatti = 0.0
    egosize=ego_size(G)
    if G.degree(v)==0: 
        return borgatti
    
    g = G.to_undirected()
    t = []    
    alters_v = egosize.get(v)
    for u in alters_v:
        alters_w = egosize.get(u)
        for w in alters_w:
            if ((w != v) and ( w!= u)):
                if t.count(g.get_edge(u,w)) == 0:
                    t.append(g.get_edge(u,w))

    borgatti =  len(egosize.get(v)) - (2*len(t)/len(egosize.get(v)))
    return borgatti 
 


def constraint(G,v): 
    """
    Burt's constraint measure (equation 2.4, page 55 of Burt, 1992). Essentially a
    measure of the extent to which v is invested in people who are invested in
    other of v's alters (neighbors).  The "constraint" is characterized
    by a lack of primary holes around each neighbor.  Formally:
    constraint(v) = sum_{w in MP(v), w != v} localConstraint(v,w)
    where MP(v) is the subset of v's neighbors that are both predecessors and successors of v.
    """
    result = 0.0
    if G.degree(v)==0: 
        return result
    ego = ego_size(G)
    alters_v = ego.get(v)
    for w in alters_v: 
        result += local_constraint(G,v,w)
    return result
    


def local_constraint(G, v1, v2):
    """
    Returns the local constraint on v from a lack of primary holes
    around its neighbor v2.  Based on Burt's equation 2.4.  Formally:
    localConstraint(v1, v2) = ( p(v1,v2) + ( sum_{w in N(v)} p(v1,w) *
    p(w, v2) ) )^2 where
    N(v) = v.getNeighbors()
    p(v,w) = normalized mutual edge weight of v and w
    """
    nmew_vw = normalized_mutual_edge_weight(G,v1,v2)
    
    inner_result = 0
    if G.directed==False:
        for w in G.neighbors_iter(v1):
            inner_result += \
                normalized_mutual_edge_weight(G,v1,w) * \
                normalized_mutual_edge_weight(G,w,v2)
    if G.directed == True:
        ego = ego_size(G)
        alters_v1 = ego.get(v1)
        for w in alters_v1:
            #if (w!= v2):
            inner_result += \
                normalized_mutual_edge_weight(G,v1,w) * \
                normalized_mutual_edge_weight(G,w,v2)
    return (nmew_vw + inner_result) * (nmew_vw + inner_result)
    


def hierarchy(G,v):
    """
    Calculates the hierarchy value for a given vertex.  Returns NaN when
    v's degree is 0, and 1 when v's degree is 1.
    Formally:
    hierarchy(v) = (sum_{v in N(v), w != v} s(v,w) * log(s(v,w))}) / (v.degree() * Math.log(v.degree())
    where
    N(v) = v.getNeighbors()
    s(v,w) = localConstraint(v,w) / (aggregateConstraint(v) / v.degree())
    """
    
    if (G.degree(v)== 0):
        return "."
    if (G.degree(v)== 1):
        return 1
    ego = ego_size(G)
    alters_v = ego.get(v)
    v_degree = len(alters_v) 
    
    
    v_constraint = aggregate_constraint(G,v)
    sl_constraint =0
    numerator = 0
    
    for w in alters_v:
        sl_constraint = local_constraint(G,v, w) /(v_constraint / v_degree)
        numerator += sl_constraint * math.log(sl_constraint)
        
    return numerator / (v_degree * math.log(v_degree))
    


def aggregate_constraint(G,v):
    """
    The aggregate constraint on v.  Based on Burt's equation 2.7.
    Formally: aggregateConstraint(v) = sum_{w in N(v)} localConstraint(v,w) * O(w)
    """
    result = 0
    egosize = ego_size(G)
    alters_u = egosize.get(v) 
    for w in alters_u: 
        result += local_constraint(G,v,w)*organizational_measure(G, w)
    return result

def organizational_measure(G, v):
    """
    A measure of the organization of individuals within the subgraph
    centered on v.  Burt's text suggests that this is
    in some sense a measure of how "replaceable" v is by
    some other element of this subgraph.  Should be a number in the
    closed interval [0,1].
    
    This implementation returns 1.  Users may wish to override this
    method in order to define their own behavior.
    """
    return 1.0
       

def mutual_weight(G,u,v,normalized=False,maximum=False):
    #w=G.get_edge(u,v,default=0.0)+G.get_edge(v,u,default=0.0)
    if G.get_edge_data(u,v):
        w=1
    else:
        w=0
    
    if normalized:
        denominator = 0.0
        #if G.directed==False:
        for n in G.neighbors_iter(u):
            denominator += mutual_weight(G,u,n)
        #if G.directed == True:
        #    egosize = ego_size(G)
        #    alters_u = egosize.get(u)
        #    for n in alters_u: 
        #        denominator += mutual_weight(G,u,n)
        w/=denominator
    if maximum:
        denominator = 0.0
        #if G.directed==False:
        for n in G.neighbors_iter(u):
            denominator = max(denominator,mutual_weight(G,u,n))
        #if G.directed == True:
        #    for n in G.predecessors_iter(u):
        #        denominator = max(denominator,mutual_weight(G,u,n))
        w/=denominator
    return w

def normalized_mutual_edge_weight(G,u,v):
    return mutual_weight(G,u,v,normalized=True)
 
def max_scaled_mutual_edge_weight(G,u,v):
    return mutual_weight(G,u,v,maximum=True)

"""
For debugging purposes
print "Node" + '\t'+ "Borgatti" +'\t'+ "Diederik" + '\t' + "Constraint" + '\t' + "Hierarchy" + '\t' + "Ego_Density"
for n in G:
    b = borgatti_effective_size(G,n)
    e = effective_size(G,n)
    c = constraint(G,n)
    h = hierarchy(G,n)
    d = ego_density(G,n)
     
    print n +'\t'+ str(b) + '\t' + str(e) + '\t' + str(c) + '\t' + str(h) + '\t' + str(d)
"""
```
