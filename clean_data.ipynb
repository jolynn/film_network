```python
#setup
%matplotlib inline
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from collections import defaultdict
```

```python
#helper functions
def is_tv(title):
    """
    Given a title, return whether the title is a TV show.
    
    Two ways to determine whether the title is TV:
    (1) The word '(TV)' in the title
    (2) The title includes season and espisode number (#n.n)
    """
    if '(TV)' in title:
        return True
    if '{' in title and '}' in title and re.search(r'#[0-9][0-9]*.[0-9][0-9]*', title):
        return True
    return False
     
def get_year(title):
    """
    Given a title, return the year of production (int). 
    If year of production is missing, return 0.
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return int(re.search(r'\d{4}', year.group(0)).group(0))
    if re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title):
        return 0
    raise Exception, 'unknown year format: '+title
         
def get_name(title):
    """
    Given a title, return name of movie/TV.
    """
    
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return title.split(year.group(0))[0]
    
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return title.split(missing.group(0))[0]
    
    raise Exception, 'unknown name format: '+title    
```

```python
def read_country(f):
    """
    """
    
    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    title_ls = []
    country_ls = []
    
    title = None
    country = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        assert len(segs) >= 2, 'wrong format: '+line
        title = segs[0]
        country = segs[-1]
        title_ls.append(title)
        country_ls.append(country)
                
    #save data to a dataframe
    df = pd.DataFrame({'title':title_ls, 'country':country_ls}, columns=['title', 'country'])
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df
```

```python
print 'There are %s movies/TVs made in the US'%country_df[country_df.country=='USA'].shape[0]
US_df = country_df[country_df.country=='USA']
```

```python
def has_episode_num(titles):
    """
    """
    for title in titles:
        if '{' in title and '}' in title and re.search(r'#[0-9][0-9]*.[0-9][0-9]*', title):
            return True
    return False

def is_potential_prob(row, df, ind):
    """
    """
    i = ind.index(row.name)
    subsequent = df.iloc[i:i+5]
    same_name_df = subsequent[subsequent.name==row['name']]
    nrow = same_name_df.shape[0]
    if nrow>1:
        return (not row['tv'] and not row['vg'] and not row['vm'] and \
                has_episode_num(same_name_df.title.values.tolist()))
    return False
```

Run the above function in some randomly selected rows on a smaller scale to see if my algorithm to re-define TV status is correct.

```python
found = False
ind = US_df.index.tolist()

while not found:

    res = pd.DataFrame(columns=['title', 'country', 'name', 'year', 'tv', 'vm', 'vg', 'potential_prob'])

    for i in range(10):
        start = random.randrange(0, US_df.shape[0]-50)
        sub_df = US_df[start:start+50]
        sub_df['potential_prob'] = sub_df.apply(lambda row: is_potential_prob(row, US_df, ind=ind), axis=1)
        res = pd.concat([res, sub_df])

    count = res.potential_prob.values.tolist().count(True)
    if count>0:
        print 'Found %s entries that potentially misidentified TV status'%count
        found = True
```

```python
#correct TV status
def is_prob(row, df, ind):
    """
    copy is_potential_prob
    add a printing statement to track progress
    """
    i = ind.index(row.name)
    if i%10000==0:
        print i
    
    subsequent = df.iloc[i:i+5]
    same_name_df = subsequent[subsequent.name==row['name']]
    nrow = same_name_df.shape[0]
    if nrow>1:
        return (not row['tv'] and not row['vg'] and not row['vm'] and \
                has_episode_num(same_name_df.title.values.tolist()))
    return False

US_df['prob'] = US_df.apply(is_prob, axis=1, df=US_df, ind=ind)

#save results to csv
US_df.to_csv('US_work_with_potential_prob.csv', index=False)
```

```python
def read_data(f, role):
    """
    Given a file path and crew role name, 
    return a dataframe that reads the data set in the file path. 
    The dataframe contains the following fields:
        person: crew member's name
        title: messy title from the dataset
        role: crew member's role 
        name: cleaned title of movie/tv
        year: year of production
        tv: is TV or not
        vm: is video movie or not
        vg: is video game or not
          
    """

    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    person_ls = []
    title_ls = []
    
    person = None
    title = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        if len(segs) >= 2:
            person = segs[0]
            title = segs[-1]
            person_ls.append(person)
            title_ls.append(title)

        elif len(segs) == 1:
            if len(segs[0]) != 0:
                title = segs[0]
                person_ls.append(person)
                title_ls.append(title)
            
        else:
            print segs  
    
    #save data to a dataframe
    df = pd.DataFrame({'person':person_ls, 'title':title_ls}, columns=['person', 'title'])
    df['role'] = role
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df
```

```python
def read_language(f):
    """
    
    Note: some movies had language 'None'
    """
    
    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    title_ls = []
    language_ls = []
    
    title = None
    language = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        assert len(segs) >= 2, 'wrong format: '+line
        title = segs[0]
        
        if segs[-1].startswith('('):
            language = segs[-2]
        else:
            language = segs[-1]
        title_ls.append(title)
        language_ls.append(language)
        
        ##I used the following codes to find invalid language field
        ##some language field lacks a parenthesis or had some random characters (i.e. English 1)
        ##I edited those fields in the txt file
        
        #if language == '1':
        #    print line
        #    return line
                
    #save data to a dataframe
    df = pd.DataFrame({'title':title_ls, 'language':language_ls}, columns=['title', 'language'])
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df  
```
