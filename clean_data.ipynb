```python
#helper functions
def is_tv(title):
    """
    Given a title, return whether the title is a TV show.
    
    Two ways to determine whether the title is TV:
    (1) The word '(TV)' in the title
    (2) The title includes season and espisode number (#n.n)
    (3) The title contains (yyyy-mm-dd)
    (4) The title contians {episode name}
    """
    if '(TV)' in title:
        return True
    if '{' in title and '}' in title and re.search(r'#[0-9][0-9]*.[0-9][0-9]*', title):
        return True
    if re.search(r'\{\([0-9]{4}-[0-9]{2}-[0-9]{2}\)\}', title):
        return True
    if re.search(r'\([0-9]{4}\)\s\{(.+)\}', title):
        return True
    return False
     
def get_year(title):
    """
    Given a title, return the year of production (int). 
    If year of production is missing, return 0.
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return int(re.search(r'\d{4}', year.group(0)).group(0))
    if re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title):
        return 0
    raise Exception, 'unknown year format: '+title
    
def get_year_id(title):
    """
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return re.search(r'\d{4}[/]*[A-Z]*', year.group(0)).group(0)
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return re.search(r'[\?]{4}[/]*[A-Z]*', missing.group(0)).group(0)
    raise Exception, 'unknown year format: '+title
         
def get_name(title):
    """
    Given a title, return name of movie/TV.
    """
    
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return title.split(year.group(0))[0]
    
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return title.split(missing.group(0))[0]
    
    raise Exception, 'unknown name format: '+title    
```

```python
def read_country(f):
    """
    """
    
    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    title_ls = []
    country_ls = []
    
    title = None
    country = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        assert len(segs) >= 2, 'wrong format: '+line
        title = segs[0]
        country = segs[-1]
        title_ls.append(title)
        country_ls.append(country)
                
    #save data to a dataframe
    df = pd.DataFrame({'title':title_ls, 'country':country_ls}, columns=['title', 'country'])
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['year_id'] = df.title.map(get_year_id)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df
```

##Read crew member datasets

```python
def read_data(f, role):
    """
    Given a file path and crew role name, 
    return a dataframe that reads the data set in the file path. 
    The dataframe contains the following fields:
        person: crew member's name
        title: messy title from the dataset
        role: crew member's role 
        name: cleaned title of movie/tv
        year: year of production
        tv: is TV or not
        vm: is video movie or not
        vg: is video game or not
          
    """

    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    person_ls = []
    title_ls = []
    
    person = None
    title = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        if len(segs) >= 2:
            person = segs[0]
            title = segs[-1]
            person_ls.append(person)
            title_ls.append(title)

        elif len(segs) == 1:
            if len(segs[0]) != 0:
                title = segs[0]
                person_ls.append(person)
                title_ls.append(title)
            
        else:
            print segs  
    
    #save data to a dataframe
    df = pd.DataFrame({'person':person_ls, 'title':title_ls}, columns=['person', 'title'])
    df['role'] = role
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['year_id'] = df.title.map(get_year_id)
    df['tv'] = df.title.map(is_tv)
    df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game
    
    return df
```

##Create bipartite network

```python
def add(movie_df, crew_df, net, role):
    """
    Add nodes and edges to the bipartite network 
    Return network.
    """
    
    movie_df['new_name'] = movie_df.apply(lambda row: row['name'].strip('"').replace(' ', '_')+'_(%s)'%row['year_id'], axis=1)
    merged = movie_df.merge(crew_df[['name', 'year_id', 'person', 'role']], how='left', on=['name', 'year_id'])
    valid_merged = merged[~merged.person.isnull()]

    #add crew members 
    crews = list(set(valid_merged.person.tolist()))
    net.add_nodes_from(crews, bipartite=0, role=role)
    
    #add movies
    unique_movies_df = valid_merged.drop_duplicates(subset=['name', 'year_id'])
    net.add_nodes_from(unique_movies_df.new_name.tolist(), bipartite=1, year=unique_movies_df.year.tolist())
    
    #add edges
    edges = [(row['new_name'], row['person']) for i, row in valid_merged.iterrows()]
    net.add_edges_from(edges)
    #valid_merged.apply(lambda row: net.add_edge((row['new_name'], row['person'])), axis=1)

    return net
```

```python
biptnet = nx.Graph()
biptnet = add(final_sample, cinmgraph_df, biptnet, 'cinematographer')
biptnet = add(final_sample, composer_df, biptnet, 'composer')
biptnet = add(final_sample, director_df, biptnet, 'director')
biptnet = add(final_sample, editor_df, biptnet, 'editor')
biptnet = add(final_sample, producer_df, biptnet, 'producer')
biptnet = add(final_sample, prod_designer_df, biptnet, 'production-designer')
biptnet = add(final_sample, writer_df, biptnet, 'writer')
```
