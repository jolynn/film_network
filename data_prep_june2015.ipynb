```python
#helper functions
def get_name_imdb(title):
    """
    Given a title, return name of movie/TV.
    """
    
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return title.split(year.group(0))[0]
    
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return title.split(missing.group(0))[0]
    
    raise Exception, 'unknown name format: '+title  
    
def get_year_imdb(title):
    """
    Given a title, return the year of production (int). 
    If year of production is missing, return 0.
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return int(re.search(r'\d{4}', year.group(0)).group(0))
    if re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title):
        return 0
    raise Exception, 'unknown year format: '+title

def is_tv_imdb(title):
    """
    Given a title, return whether the title is a TV show.
    
    Two ways to determine whether the title is TV:
    (1) The word '(TV)' in the title
    (2) The title includes season and espisode number (#n.n)
    (3) The title contains (yyyy-mm-dd)
    (4) The title contians {episode name}
    """
    if '(TV)' in title:
        return True
    if '{' in title and '}' in title and re.search(r'#[0-9][0-9]*.[0-9][0-9]*', title):
        return True
    if re.search(r'\{\([0-9]{4}-[0-9]{2}-[0-9]{2}\)\}', title):
        return True
    if re.search(r'\([0-9]{4}[/]*[I]*\)\s\{(.+)\}', title):
        return True
    return False
     
def get_year_id_imdb(title):
    """
    Given a title, return the year id. 
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return re.search(r'\d{4}[/]*[A-Z]*', year.group(0)).group(0)
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return re.search(r'[\?]{4}[/]*[A-Z]*', missing.group(0)).group(0)
    raise Exception, 'unknown year format: '+title
```

```python
def read_country(f):
    """
    Given a file path f, return a dataframe constructed from the file f. 
    """
    
    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    title_ls = []
    country_ls = []
    
    title = None
    country = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        assert len(segs) >= 2, 'wrong format: '+line
        title = segs[0]
        country = segs[-1]
        title_ls.append(title)
        country_ls.append(country)
                
    #save data to a dataframe
    df = pd.DataFrame({'title_imdb':title_ls, 'country':country_ls}, columns=['title_imdb', 'country']) #avoid repetitive name during merge
    
    #extract more information from title
    df['name'] = df.title_imdb.map(get_name_imdb)
    df['year'] = df.title_imdb.map(get_year_imdb)
    df['year_id'] = df.title_imdb.map(get_year_id_imdb)
    df['tv'] = df.title_imdb.map(is_tv_imdb)
    df['vm'] = df.title_imdb.map(lambda x: '(V)' in x) #video movie
    df['vg'] = df.title_imdb.map(lambda x: '(VG)' in x) #video game
    
    return df
```

```python
#helper functions
def get_name(title):
    """
    Given a title, return name of movie/TV.
    """
    
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return title.split(year.group(0))[0]
    
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return title.split(missing.group(0))[0]
    
    raise Exception, 'unknown name format: '+title  
    
def get_year(title):
    """
    Given a title, return the year of production (int). 
    If year of production is missing, return 0.
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return int(re.search(r'\d{4}', year.group(0)).group(0))
    if re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title):
        return 0
    raise Exception, 'unknown year format: '+title
     
def get_year_id(title):
    """
    Given a title, return the year id. 
    """
    year = re.search(r'\s\(\d{4}[/]*[A-Z]*\)', title)
    if year:
        return re.search(r'\d{4}[/]*[A-Z]*', year.group(0)).group(0)
    missing = re.search(r'\s\([\?]{4}[/]*[A-Z]*\)', title)
    if missing:
        return re.search(r'[\?]{4}[/]*[A-Z]*', missing.group(0)).group(0)
    raise Exception, 'unknown year format: '+title

def read_data(f, role):
    """
    Given a file path and crew role name, 
    return a dataframe that reads the data set in the file path. 
    The dataframe contains the following fields:
        person: crew member's name
        title: messy title from the dataset
        role: crew member's role 
        name: cleaned title of movie/tv
        year: year of production
    """

    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    person_ls = []
    title_ls = []
    
    person = None
    title = None
    
    for line in lines:
        line = line.strip()
        segs = line.split('\t')
        
        if len(segs) >= 2:
            person = segs[0]
            title = segs[-1]
            person_ls.append(person)
            title_ls.append(title)

        elif len(segs) == 1:
            if len(segs[0]) != 0:
                title = segs[0]
                person_ls.append(person)
                title_ls.append(title)
            
        else:
            print segs  
    
    #save data to a dataframe
    df = pd.DataFrame({'person':person_ls, 'title':title_ls}, columns=['person', 'title'])
    df['role'] = role
    
    #extract more information from title
    df['name'] = df.title.map(get_name)
    df['year'] = df.title.map(get_year)
    df['year_id'] = df.title.map(get_year_id)
    #df['tv'] = df.title.map(is_tv)
    #df['vm'] = df.title.map(lambda x: '(V)' in x) #video movie
    #df['vg'] = df.title.map(lambda x: '(VG)' in x) #video game

    return df
```

##Create bipartite network

```python
def add(net, movie_df, crew_df, role):
    """
    Add nodes and edges to the bipartite network.
    Return the modified network.
    """

    #add crew members 
    crews = list(set(crew_df.person.tolist()))
    net.add_nodes_from(crews, bipartite=0, role=role)
    
    #add movies
    movie_nodes = [(row['title_imdb'], dict(year=row['year'])) for i, row in movie_df.iterrows()]
    net.add_nodes_from(movie_nodes, bipartite=1)
    
    #add edges
    edges = [(row['title_imdb'], row['person']) for i, row in crew_df.iterrows()]
    net.add_edges_from(edges)
    #valid_merged.apply(lambda row: net.add_edge((row['title_imdb'], row['person'])), axis=1)

    return net
```

```python
biptnet = nx.Graph()
biptnet = add(biptnet, sample_df, crew_df[crew_df.role=='cinematographer'], 'cinematographer')
biptnet = add(biptnet, sample_df, crew_df[crew_df.role=='composer'], 'composer')
biptnet = add(biptnet, sample_df, crew_df[crew_df.role=='director'], 'director')
biptnet = add(biptnet, sample_df, crew_df[crew_df.role=='editor'], 'editor')
biptnet = add(biptnet, sample_df, crew_df[crew_df.role=='producer'], 'producer')
biptnet = add(biptnet, sample_df, crew_df[crew_df.role=='production-designer'], 'production-designer')
biptnet = add(biptnet, sample_df, crew_df[crew_df.role=='writer'], 'writer')
```

```python
def get_film_connections(title):
    '''Given a film title, return a list of films made within the past three years that share crew with the given film'''
    
    y = get_year_imdb(title)
    neighbors = film_net.neighbors(title)
    recent_n = [n for n in neighbors if y-get_year_imdb(n)>0 and y-get_year_imdb(n)<=3]
    return len(recent_n)

sample_df['connect'] = sample_df.title_imdb.map(get_film_connections)
sample_df.head()
```

###Create whole crew network over time

There is a separate network for each year. The network contains ties of people who worked on some films within the past three years. It could also contain ties from more than three years ago if the ties are active over time. For example, if a movie was made in 2000, and two crew members worked together in 1998 and 1996. Then in the whole network of 2000, the two crew members have a tie with weight two (they worked on two movies in the past). 

```python
def get_weight(co_movies_years, yr, window=3):
    '''A recursive function:
    Given a list of tuples (movie title, year), year in which the network is constructed, and window of active ties, 
    return the number of active ties from the given year to three (or the given window) years before.'''
    
    if len(co_movies_years)==0:
        return 0
    else:
        item = co_movies_years.pop(0)
        if yr-item[1] > window-1: #the film was made three (or the given window) years before the given year
            return 0
        else:
            return 1+get_weight(co_movies_years, item[1], window)
```

```python
def convert_one_mode(biptnet, yr, window=3):
    '''
    Convert bipartite network to one-mode network of collaboration of the given yr, 
    and add an edge attribute called weight (number of active ties within three years). 
    '''
    
    #movies made before the given year
    movies = [m for m, d in biptnet.nodes(data=True) if (d['bipartite']==1 and d['year'] in range(1991, yr+1))]
    
    crews = []
    for m in movies: crews += biptnet.neighbors(m)
        
    sub_net = nx.subgraph(biptnet, movies+crews)
    
    #convert to one-mode network
    one_mode = bipartite.projected_graph(sub_net, crews)
    
    #calculate weight and remove nodes and edges with weight zero
    edge_wt = {}
    nodes = []
    for (p1, p2) in one_mode.edges():
        co_movies = list(set(sub_net.neighbors(p1))&set(sub_net.neighbors(p2)))
        co_movies_years = [(m, get_year(m)) for m in co_movies]
        co_movies_years = sorted(co_movies_years, key=itemgetter(1), reverse=True)
        wt = get_weight(co_movies_years, yr, window)
        if wt>0:
            edge_wt[(p1, p2)] = wt
            nodes.append(p1)
            nodes.append(p2)
    
    #this way, we won't miss isolated nodes
    for m, d in biptnet.nodes(data=True):
        if (d['bipartite']==1 and d['year'] in range(yr-window+1, yr+1)):
            nodes += biptnet.neighbors(m)
            
    #remove duplicated nodes
    nodes = list(set(nodes))
    
    #create one-mode network
    subgraph = nx.Graph()
    subgraph.add_nodes_from([(n, d) for (n, d) in one_mode.nodes(data=True) if n in nodes])
    subgraph.add_edges_from(edge_wt.keys())
    nx.set_edge_attributes(subgraph, 'wt', edge_wt)
    
    print 'year:', yr, 
    print '# nodes in all movies:', len(one_mode.nodes()), 
    print '# nodes in network:', len(subgraph.nodes())
    
    return subgraph
```

###Within-team density

Calculated based on the collaboration network up until the year before the current movie

```python
def get_density(movie):
    '''
    Given a movie title, return unweighted density, weighted density, 
    and number of active crew members from the year before.
    '''
    
    last_yr = get_year(movie)-1
    net_last_yr = crew_networks[last_yr]
    
    crews = biptnet.neighbors(movie)
    crews_last_yr = list(set(crews)&set(net_last_yr.nodes()))
    movie_net_last_yr = net_last_yr.subgraph(crews_last_yr)
    
    try:
        #calculate densities
        n = len(crews) #number of nodes
        if n<=1:
            return 0, 0, len(crews_last_yr)
        else:
            den = 2.0*len(movie_net_last_yr.edges())/n/(n-1)
            wt_den = 2.0*sum(nx.get_edge_attributes(movie_net_last_yr, 'wt').values())/n/(n-1)
            return den, wt_den, len(crews_last_yr)
    except nx.NetworkXError:
        print 'error'
        return None, None, None
```

```python
#didn't figure out how to calculate continuous entropy. use two groups cut by mean for now. 

def get_bt_centrality_entropy(movie):
    '''
    Given a film title, return the entropy of betweenness centrality of crew members
    based on the collaboration network the year before the movie was made
    '''
    
    last_yr = get_year(movie)-1
    crews = biptnet.neighbors(movie)
    crews_bt = []
    cutpoint = np.mean(bt_centrality[last_yr].values())
    for c in crews:
        if bt_centrality[last_yr].has_key(c):
            if bt_centrality[last_yr][c]>= cutpoint:
                crews_bt.append(1)
            else:
                crews_bt.append(0)
        else: #the crew member wasn't active the year before
            crews_bt.append(-1)
    
    props = []
    for i in range(-1, 2):
        props.append(sum([bt==i for bt in crews_bt])/float(len(crews)))
        
    return scipy.stats.entropy(props)
```

```python
def readbiz(f):
    #read lines from text file
    f = open(f, 'r')
    lines = f.readlines()
    f.close()
    
    #iterate through each line and extract crew member name and movie title
    fact_ls = []
    facttype_ls = []
    title_ls = []
    
    fact = None
    title = None
    facttype = None
    
    for line in lines:
        line = line.strip()
        if line == '' or line[0:5] == '-----':
            continue
        else:
            facttype = line[0:2]
            fact = line[4:]

            if facttype == "MV":
                title = fact
            else:
                fact_ls.append(fact)
                facttype_ls.append(facttype)
                title_ls.append(title)
            
    df = pd.DataFrame({ 'title_imdb':title_ls,'facttype':facttype_ls,'fact':fact_ls,}, columns=['title_imdb','facttype','fact'])
    return df

bizdf = readbiz('business.txt') 
bizdf.facttype.unique() #Looks right. WG isn't in the key in business.list, but it does appear in the data as a category
budget_df = bizdf[bizdf.facttype == 'BT']
budget_df['name'] = budget_df.title_imdb.map(get_name)
budget_df['year'] = budget_df.title_imdb.map(get_year)
budget_df['year_id'] = budget_df.title_imdb.map(get_year_id)
budget_df['edited_name'] = budget_df.apply(lambda row: row['name'].strip('"').replace(' ', '_')+'_(%s)'%row['year_id'], axis=1)
budget_df['currency'] = budget_df['fact'].apply(lambda x: x.split()[0])
budget_df['budget_biz'] = budget_df['fact'].apply(lambda x: int(x.split()[1].replace(',','').replace('.','')))
budget_df.head()
```
